```python
import pandas as pd 
from mlxtend.frequent_patterns import apriori , association_rules 
```

```python
address = r'C:\Users\Dell\Python For Data Science\Data\groceries.csv'

data = pd.read_csv(address )
```

```python
data.head()
```

```python
basket_sets = pd.get_dummies(data)
```

```python
basket_sets.head()
```

```python
x = apriori(basket_sets , min_support = 0.02)
```

```python
y = apriori(basket_sets , min_support = 0.02 , use_colnames = True)
```

```python
df = basket_sets

frequent_itemsets = apriori(basket_sets , min_support = 0.002 , use_colnames = True)

frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x : len(x))

frequent_itemsets.head(100)
```

```python
frequent_itemsets[frequent_itemsets['length'] >= 3]
```

```python
rules = association_rules(frequent_itemsets , metric = "confidence" , min_threshold = 0.5)
rules.head()
```

```python
rules = association_rules(frequent_itemsets , metric = "lift" , min_threshold = 1)
rules.head()
```

```python
round(rules[(rules['lift']>= 5) & (rules['confidence']>= 0.5)] , 4)
```
